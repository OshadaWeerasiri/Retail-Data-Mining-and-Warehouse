### **Predict Shipment Status**

```python
query = """
SELECT
    f.Discount,
    f.Quantity,
    f.Sales,
    f.Profit,
    s.ShipMode,
    s.ShipStatus
FROM Fact_Sales f
JOIN Dim_Shipment s
  ON f.ShipID = s.ShipID
"""

df2 = pd.read_sql(query, conn)
```

```python
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_mode = LabelEncoder()
le_status = LabelEncoder()

df2['ShipMode_Code'] = le_mode.fit_transform(df2['SHIPMODE'])
df2['ShipStatus_Code'] = le_status.fit_transform(df2['SHIPSTATUS'])
```

```python
df2.head(10)
```

### **Predict Sales Forecast**

```python
query2 = """
SELECT
    f.Discount,
    f.Quantity,
    f.Sales,
    f.Profit,
    f.ProfitRatio,
    f.SalesForecast,
    p.Category,
    p.Sub_Category,
    c.Segment,
    c.Region,
    c.City,
    c.State
FROM Fact_Sales f
JOIN Dim_Shipment s ON f.ShipID = s.ShipID
JOIN Dim_Product p ON f.ProductID = p.ProductID
JOIN Dim_Customer c ON f.CustomerID = c.CustomerID
WHERE f.SalesForecast IS NOT NULL
"""
df3 = pd.read_sql(query2, conn)
df3.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_mode = LabelEncoder()
le_status = LabelEncoder()

df3['Category'] = le_mode.fit_transform(df3['CATEGORY'])
df3['Sub_Category'] = le_status.fit_transform(df3['SUB_CATEGORY'])
df3['Segment'] = le_mode.fit_transform(df3['SEGMENT'])
df3['Region'] = le_status.fit_transform(df3['REGION'])
df3['City'] = le_status.fit_transform(df3['CITY'])
df3['State'] = le_mode.fit_transform(df3['STATE'])
```

### **Cluster Orders by Buying Behavior**

```python
query4 = """
SELECT
    c.CustomerID,
    c.Segment,
    c.Country,
    AVG(f.Quantity) AS AvgQuantity,
    AVG(f.Discount) AS AvgDiscount,
    AVG(f.Sales) AS AvgSales,
    AVG(f.Profit) AS AvgProfit,
    AVG(f.DaystoShipActual) AS AvgShipTime,
    s.ShipMode
FROM Fact_Sales f
JOIN Dim_Customer c ON f.CustomerID = c.CustomerID
JOIN Dim_Shipment s ON f.ShipID = s.ShipID
GROUP BY c.CustomerID, c.Segment, c.Country, s.ShipMode
"""
```

```python
df5 = pd.read_sql(query4, conn)
df5.head()
```

```python
# One-hot encode categorical features
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Assume df is your dataframe
categorical = ['SEGMENT', 'COUNTRY', 'SHIPMODE']
numerical = ['AVGQUANTITY', 'AVGDISCOUNT', 'AVGSALES', 'AVGPROFIT', 'AVGSHIPTIME']

# Preprocessing
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical),
    ('cat', OneHotEncoder(), categorical)
])

# Build clustering pipeline
pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('cluster', KMeans(n_clusters=4, random_state=42))
])

# Fit and predict clusters
df5['Cluster'] = pipeline.fit_predict(df5)
```

```python

print(df5['Cluster'].value_counts())

# Visualize with PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
reduced = pca.fit_transform(pipeline.named_steps['preprocess'].transform(df5))

plt.figure(figsize=(8, 6))
plt.scatter(reduced[:, 0], reduced[:, 1], c=df5['Cluster'], cmap='viridis')
plt.title('Customer Clusters by Buying & Shipping Behavior')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
query5 = """
SELECT
    c.City,
    c.State,
    c.Country,
    c.Latitude,
    c.Longitude,
    f.DaystoShipActual,
    f.DaystoShipScheduled,
    (f.DaystoShipActual - f.DaystoShipScheduled) AS DeliveryDelay,
    s.ShipMode,
    s.ShipStatus
FROM Fact_Sales f
JOIN Dim_Customer c ON f.CustomerID = c.CustomerID
JOIN Dim_Shipment s ON f.ShipID = s.ShipID
WHERE c.Latitude IS NOT NULL AND c.Longitude IS NOT NULL

"""
```

```python
import folium
from folium.plugins import MarkerCluster

# Create base map
m = folium.Map(location=[df6['LATITUDE'].mean(), df6['LONGITUDE'].mean()], zoom_start=4)
marker_cluster = MarkerCluster().add_to(m)

# Add data points
for i, row in df6.iterrows():
    folium.Marker(
        location=[row['LATITUDE'], row['LONGITUDE']],
        popup=f"{row['CITY']}, Delay: {row['DELIVERYDELAY']} days, Status: {row['SHIPSTATUS']}",
        icon=folium.Icon(color="green" if row['DELIVERYDELAY'] <= 0 else "red")
    ).add_to(marker_cluster)

m.save("shipment_map.html")
```
